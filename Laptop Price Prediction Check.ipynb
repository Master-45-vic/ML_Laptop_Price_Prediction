```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt 
import warnings
import math
warnings.filterwarnings("ignore")
```

```python
import joblib

model = joblib.load("xgb_price_model.pkl")
scaler = joblib.load("scaler.pkl")
encoder = joblib.load("encoder.pkl")
```

```python
encoder.feature_names_in_
```

```python
cat_cols = ['Brand', 'Model', 'CPU', 'GPU', 'Storage type', 'Touch', 'Status']
num_cols = ['RAM', 'Storage', 'Screen']
```

```python
import numpy as np

def predict_laptop_price(input_df):

    # Select categorical & numeric
    X_cat = input_df[cat_cols]
    X_num = input_df[num_cols]

    #Filling Missing Value
    input_df["GPU"]=input_df["GPU"].fillna("Others")
    

    # Encode categorical
    X_cat_enc = encoder.transform(X_cat)

    # Scale numeric
    X_num_scaled = scaler.transform(X_num)

    # Combine
    X_final = np.hstack([X_cat_enc, X_num_scaled])

    # Covert Result back from log into Original Value 
    price = math.floor(np.expm1(model.predict(X_final)))

    return price
```

```python
Sample_data=pd.DataFrame([{'Status':'New','Brand':'HP','Model':'EliteBook','CPU':'Intel Core i7','RAM':16,'Storage':512,'Storage type':'SSD','GPU':np.nan,'Screen':14,'Touch':'No'}])
```

```python
print(predict_laptop_price(Sample_data))
```
